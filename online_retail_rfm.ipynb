# Görev 2

### Adım 1
* Recency : Müşterinin analiz tarihinden en son kaç gün önce alışveriş yaptığı
* Frequency : Müşterinin kaç adet alışveriş yaptığı
* Monetary : Müşterinin toplam bıraktığı para 

```python
df['InvoiceDate'].max()
```

```python
analys_date = dt.datetime(2011, 12, 11)
analys_date
```

### Adım 2 ve Adım 3

```python
rfm = df.groupby('Customer ID').agg({'InvoiceDate': lambda date: (analys_date - date.max()).days,
                               'Invoice': lambda x: x.nunique(),
                               'TotalPrice': lambda x: x.sum()})
rfm
```

### Adım 4

```python
rfm.columns = ['recency', 'frequency', 'monetary']
rfm
```

# Görev 3

### Adım 1 ve Adım 2

```python
rfm['recency_score'] = pd.qcut(rfm['recency'], 5, labels=[5,4,3,2,1])
rfm
```

```python
rfm['frequency_score'] = pd.qcut(rfm['frequency'].rank(method='first'), 5, labels=[1,2,3,4,5])
rfm
```

```python
rfm['monetary_score'] = pd.qcut(rfm['monetary'], 5, labels=[1,2,3,4,5])
rfm
```

```python
rfm.describe().T
```

### Adım 3

```python
rfm['RF_Score'] = (rfm['recency_score'].astype(str) + rfm['frequency_score'].astype(str))
rfm
```

# Görev 4

```python
# SEGMENTLERIN ISIMLENDIRILMESI
seg_map = {
    r'[1-2][1-2]': 'hibernating',
    r'[1-2][3-4]': 'at_risk',
    r'[1-2]5': 'cant_loose',
    r'3[1-2]': 'about_to_sleep',
    r'33': 'need_attention',
    r'[3-4][4-5]': 'loyal_customers',
    r'41': 'promising',
    r'51': 'new_customers',
    r'[4-5][2-3]': 'potential_loyalists',
    r'5[4-5]': 'champions'
}     
```

```python
rfm['segment'] = rfm['RF_Score'].replace(seg_map, regex=True)
rfm
```

# Görev 5

```python
segmentsx = rfm["segment"].value_counts().sort_values(ascending=False)
segmentsx
```

```python
import matplotlib.pyplot as plt 
import squarify
fig = plt.gcf()
ax = fig.add_subplot()
fig.set_size_inches(15, 11)

squarify.plot(sizes=segmentsx,
                label=['hibernating',
                    'at_risk',
                    'cant_loose',
                    'about_to_sleep',
                    'need_attention',
                    'loyal_customers',
                    'promising',
                    'new_customers',
                    'potential_loyalists',
                    'champions'],color=["red","yellow","blue", "green","orange"],pad=True,
            bar_kwargs={'alpha':.70}, text_kwargs={'fontsize':12})
plt.title("Müşteri Segmentasyonu",fontsize=20)
plt.xlabel('Frequency', fontsize=20)
plt.ylabel('Recency', fontsize=20)
plt.show()
```
